```python
# 브라우저를 모달로 이동
# 스크롤
# Switch to iFrame and then try to scroll down

browser.find_element(By.CLASS_NAME, 'fysCi').click()
modal = browser.find_element(By.CLASS_NAME, 'fysCi')

import time
#내가 필요한 만큼 스크롤
for i in range(3):
    
    try: # 예외처리 (try - except)
        browser.execute_script('arguments[0].scrollTo(0, arguments[0].scrollHeight);', modal)
        time.sleep(2)
        #print(browser.execute_script('arguments[0].scrollTo(0, arguments[0].scrollHeight);', modal))
        #print 못씀! > 동작
    except:
        print("try에 오류가 났을 때 여기로 오게 됩니다.")
    
```

```python
import pandas as pd

containers = browser.find_elements(By.CLASS_NAME, 'RHo1pe')

# data_list 생성 : 반복문 안에 있으면 계속 초기화됨(빈 리스트이기 때문)
data_list = []

# 클래스 밑에 attribute
for container in containers:
    user_name = container.find_element(By.CLASS_NAME, 'X5PpBb').text
    
    # 10번째 인덱스에서 자르기
    rating = container.find_element(By.CLASS_NAME, 'iXRFPc').get_attribute('aria-label')[10]
    
    date = container.find_element(By.CLASS_NAME, 'bp9Aid').text
    
    content = container.find_element(By.CLASS_NAME, 'h3YV2d').text
#     print("==="*10)
#     print(content)
    
    #dict 형태로 데이터를 만든다. dataframe에 들어가려면 list형태로 만들어야 한다.
    data = {
        "user_name": user_name,
        "rating": rating,
        "date": date,
        "content":content
    }
    
    data_list.append(data)
    
df = pd.DataFrame(data_list)
df.to_csv('./palystore_comment.csv', encoding = 'utf-8-sig') #csv파일로 저장
```
