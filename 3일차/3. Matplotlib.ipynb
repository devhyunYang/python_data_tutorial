```python
import matplotlib.pyplot as plt
import numpy as np
```

# 기본 그래프

```python
plt.plot([1,2,3,4,5,6,5,4,3,2,1])
plt.show()
```

```python
## sin 함수 그려보기
x = np.arange(0, 12, 0.01) # 0~12까지 0.01간격
y = np.sin(x)
```

```python
plt.plot(x,y) # plot 그래프 그리기
plt.show()
```

## 타이틀 설정

```python
plt.title("sin graph")
plt.plot(x,y) # plot 그래프 그리기
plt.show()
```

## Grid 설정

```python
plt.plot(x,y) # plot 그래프 그리기
plt.grid()
plt.show()
```

## x,y 라벨

```python
plt.plot(x,y) # plot 그래프 그리기
plt.xlabel('time')
plt.ylabel('Amplitude') #진폭
plt.show()
```

```python
## 범례(legend)
plt.plot(x,y) # plot 그래프 그리기
plt.plot(x,y-1)
plt.xlabel('time')
plt.ylabel('Amplitude') #진폭
# plt.legend()
plt.legend(loc=(1.05, 0.05))
plt.show()
```
